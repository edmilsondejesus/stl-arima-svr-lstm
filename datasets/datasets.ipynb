```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
# load Lynx Canadian dataset
ds = pd.read_csv('lynx.csv', usecols=[1,2],engine='python')
ds['time'].astype(int)
ds['value'].astype(int)
plt.plot(ds['time'].values, ds['value'].values)
plt.xlabel('year')
plt.ylabel('amount')
plt.savefig("Fig1Lynx.eps", format="eps", transparent=False)
plt.show()
```
