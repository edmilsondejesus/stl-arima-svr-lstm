```python
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt

# Carregue os dados das manchas solares (exemplo)
# Substitua isso pelo seu próprio conjunto de dados
data = pd.read_csv("..\datasets\SN_m_tot_V2.0.csv")
data.columns = ["ds", "y"]  # Prophet espera colunas com nomes específicos
```

```python

# Crie um modelo Prophet
model = Prophet()

# Ajuste o modelo aos dados
model.fit(data)

# Crie um dataframe para as datas futuras que você deseja prever
future = model.make_future_dataframe(periods=365)  # Previsão para um ano (365 dias)

# Faça a previsão
forecast = model.predict(future)

# Visualize os resultados
fig = model.plot(forecast)
plt.title("Previsão de Manchas Solares")
plt.xlabel("Data")
plt.ylabel("Manchas Solares")
plt.show()
```
